# Read airtable sections

```python
from airhockey.read import query
```

## "Soft" Queries 
Queries from `FlyGenotype` or `Stocks` (i.e. any class whose Parent is `GenotypeTables`):
- Will pull any entry where the values contains the search criteria (as opposed to having to match it exactly).
- Are case sensitive

```python
query.FlyGenotype(chr1='w',chr2='If')
```

```python
query.FlyGenotype(chr1='w',chr2='IF') #Demonstrating case-sensitivity
```

```python
query.Stocks(genotype_id=581) 
```

## "Hard" Queries 
Queries made using `TwoPhotonRecording`, or `FlySubject` 
- Searches for exact matches
- Are case sensitive

```python
query.TwoPhotonRecording()
```

```python
query.FlySubject()
```

```python
query.FlySubject(subject_id = 2501)
```

```python
query.TwoPhotonRecording(subject_id = 2501)
```
